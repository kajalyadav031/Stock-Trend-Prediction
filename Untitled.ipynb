```python
#Splitting the data
training_data=pd.DataFrame(stock_shareprice_data['Close'][0:int(len(stock_shareprice_data)*0.70)])
testing_data=pd.DataFrame(stock_shareprice_data['Close'][int(len(stock_shareprice_data)*0.70):int(len(stock_shareprice_data))])
```

```python
print(training_data.shape)
print(testing_data.shape)
```

```python
training_data.head()
```

```python
testing_data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler(feature_range=(0,1))
```

```python
training_data_array=scaler.fit_transform(training_data).reshape(-1,1)
training_data_array
```

```python
testing_data_array=scaler.fit_transform(testing_data)
testing_data_array
```

```python
training_data.shape
```

```python
testing_data.shape
```

```python
x_train=[]
y_train=[]
for i in range (100,training_data.shape[0]):
    x_train.append(training_data_array[i-100:i])
    y_train.append(training_data_array[i:0])
x_train=np.array(x_train)
```

```python
y_train=np.array(y_train)
```

```python
#ML Model
```

```python
from keras import Sequential
from tensorflow.keras.layers import Dense,Dropout,LSTM
```

```python
model=Sequential()
```

```python
x_train.shape  
#training step=50
```

```python
y_train.shape
```

```python
#input_shape is required only for the first layer 
#Higher dropout values, such as 0.5 or 0.6, can be used for more complex models
#units in a layer determines the complexity and capacity of the layer 
```

```python
model.add(LSTM(units=50,activation='relu',return_sequences=True,input_shape=(x_train.shape[1],1)))
model.add(Dropout(0.2))
model.add(LSTM(units=60,activation='relu',return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(units=80,activation='relu',return_sequences=True))
model.add(Dropout(0.4))
model.add(LSTM(units=120,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=1))          #unit=1 since we are taking only closing value
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_absolute_error')
model.fit(x_train, y_train,epochs=50)

```

```python
model.save('keras_model.h5')
```
